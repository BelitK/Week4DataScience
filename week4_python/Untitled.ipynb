```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data_tur = pd.read_csv('covid_19_data_tr.csv') # read turkey covid cases
```

```python
data_de = pd.read_csv('covid_de.csv')# read germany covid cases
```

```python
tur_confirm = data_tur.iloc[0:,2].values #confirmed cases tr
tur_death = data_tur.iloc[0:,3].values # death cases tr
tur_recover = data_tur.iloc[0:,4].values # recovered cases tr
de_confirm = data_de.iloc[0:,2].values
de_confirm
```

```python
tur_Date = data_tur.iloc[0:,5].values #dates
```

```python

plt.figure(figsize=(50,15))

ax = plt.axes() 
ax.grid(linewidth=0.4, color='#8f8f8f') 

ax.set_facecolor("black")  
ax.set_xlabel('\nDate',size=25,color='#4bb4f2') 
ax.set_ylabel('Number of Confirmed Cases\n', 
              size=25,color='#4bb4f2') 
ax.set_xlim([0,60])
ax.plot(tur_Date,tur_confirm, 
        color='#1F77B4', 
        marker='o', 
        linewidth=4, 
        markersize=10, 
        markeredgecolor='#035E9B',
        label = 'confirmed cases')
ax.plot(tur_Date,tur_death, 
        color='#ffffff', 
        marker='o', 
        linewidth=4, 
        markersize=10, 
        markeredgecolor='#ffffff',
        label='Death cases')
ax.legend()
```
